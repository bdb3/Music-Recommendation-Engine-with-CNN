```python
import pandas as pd
df = pd.read_csv(r'C:\Users\Danny\Documents\spring18\DataScience\FinalProject\music.csv')
```

```python
list(df)
```

```python
#remove unnecessary columns for grouping
dfCols = list(df)
del dfCols[1] #artist.id
del dfCols[1] #artist.name
del dfCols[13] #location
del dfCols[2] #artist.mbtags.count
del dfCols[16] #release.id
del dfCols[16] #release.name
del dfCols[17] #song.id
del dfCols[25] #title
del dfCols[1]
dfCols
```

```python
allData = df[dfCols]
#remove onehotencoding features
del dfCols[8] #key
del dfCols[12] #mode - doesn't need normalization bc it is binary
del dfCols[18] #terms
dfCols
```

```python
finalizedDf = df[dfCols]
#fill nan values with mean of the column
finalizedDf = finalizedDf.fillna(finalizedDf.mean())
#fill 0 values in year column with average
###UNFINISHED#####
```

```python
from sklearn import preprocessing
normalized_df =pd.DataFrame()
for r in range(len(dfCols)):
    #new df is given normalized columns for each r value in range
    normalized_col = preprocessing.scale(finalizedDf[dfCols[r]].values)
    normalized_df[dfCols[r]] = normalized_col
```

```python
oh_k =pd.get_dummies(allData['key'])
oh_t =pd.get_dummies(allData['terms'])
#add onehotencoded columns to normalized feature set
normalized_df = normalized_df.join(oh_k)
normalized_df = normalized_df.join(oh_t)
normalized_df = normalized_df.join(allData['mode'])
normalized_df.head()
```

```python
z = [10,50,100,400,500]
for i in range(len(z)):
    from sklearn.cluster import KMeans
    kmeans = KMeans(n_clusters=z[i], random_state=0).fit(normalized_df)
    labels = kmeans.labels_
    #Glue back to originaal data
    df_final = df.copy(deep = True)
    df_final['clusters'] = labels
    df_final.to_csv("C:/Users/Danny/Documents/spring18/DataScience/FinalProject/"+ str(z[i])+ "clusters.csv", encoding='utf-8', index=False)

```
